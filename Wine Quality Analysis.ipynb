## Is a certain type of wine (red or white) associated with higher quality?

```python
wine.groupby('colour').mean()
```

```python
colour_means = wine.groupby('colour')['quality'].mean()

colour_means.plot(kind = 'bar', color = ['red', 'green'], alpha = 0.5 ,title  = 'Average wine quality by colour\n', fontsize = 15, grid = True, )

plt.xlabel('Colour',  fontsize = 20);

plt.ylabel('Quality\n', fontsize = 20);
```

```python
counts_ph.plot(kind = 'bar', color = ['red', 'green'], title = ' Counts by wine quality and colour\n');

plt.xlabel('Quality and Colour', fontsize = 18);

plt.ylabel('Count', fontsize = 18);
```

```python
bin_edges = [2.720000, 3.110000, 3.210000, 3.320000, 4.010000]

bin_names = ['High', 'Moderately High', 'Medium', 'Low']
```

```python
wine['acidity_levels'] = pd.cut(wine['pH'], bin_edges, labels=bin_names)
```

```python
wine.head()
```

## What level of acidity (pH value) receives the highest average rating?

```python
wine.groupby('acidity_levels')['pH'].mean()
```

## Do sweeter wines receive better ratings?
